```python
import pandas as pd
import numpy as np
train = pd.read_csv('datas/train.csv')
test = pd.read_csv('datas/test.csv')
```

```python
df = pd.concat([test.assign(ind="test"), train.assign(ind="train")])
df
```

```python
df.drop('Alley', inplace=True, axis=1) 
```

```python
df
```

```python
# I have converted all the object into category
for label in df.keys():
    if df[label].dtype == 'O':
        df[label] = df[label].astype('category')
```

```python
# I added 1 to all the cat codes 
for label in df.keys():
    if df[label].dtype == 'category':
        df[label] = df[label].cat.codes + 1
```

```python
# Created an dic for see which column has Null value and what's their type
def checking_null():
    info_dic = {}
    for label, column in df.items():
        if column.isna().sum() > 0:
            info_dic[label] = column.isna().sum()
            info_dic[label + '_dtype'] = df[label].dtype

    return info_dic
```

```python
checking_null()
```

```python
# I have created another column for is null property
for label in df.keys(): 
    if df[label].isna().sum() > 0:
        df[label + '_ismissing'] = df[label].isnull()
        
```

```python
df.info()
```

```python
df['LotFrontage_ismissing'].value_counts()
```

```python
# I filled all the numerical values
for label in df.keys(): 
    if df[label].isna().sum() > 0:
        mean = df[label].mean()
        df[label] = df[label].fillna(mean)
```

```python
checking_null()
```

```python
# Convert bool into numerical type
for label in df.keys():
    if df[label].dtype == bool:
        df[label] = df[label].astype(int)
```

```python
df.info()
```

```python
test, train = df[df["ind"].eq(1)], df[df["ind"].eq(2)]
```

```python
from sklearn.model_selection import train_test_split

X = train.drop('SalePrice', axis=1)
y = train.SalePrice

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge

# Creating models
models = {'LinearRegression': LinearRegression(),
          'RandomForestRegressor': RandomForestRegressor(),
          'Ridge': Ridge()}
```

```python
from sklearn.metrics import root_mean_squared_error

def fit_and_score(models):
    model_scores = {}
    for label, model in models.items(): 
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        model_scores[label] = np.sqrt(root_mean_squared_error(y_test, y_pred))

    return model_scores
```

```python
model1 = LinearRegression()
model1.fit(X_train, y_train)
asd = model1.predict(X_test)
root_mean_squared_error(y_test, asd)
```

```python
scores = fit_and_score(models)
```
